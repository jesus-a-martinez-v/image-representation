# Image representation

In this notebook we'll explore how images are represented in a computer. With that purpose in mind, we'll use libraries specially designed to handle images, matrices, linear algebra and multidimensional arrays of numbers, such as: `numpy`, `opencv`, `matplotlib`, among others.

## Importing the resources

First, we must import the libraries we'll need.

```python
import numpy as np
import matplotlib.image as mpimg  # Read images

import matplotlib.pyplot as plt
import cv2  # OpenCV, a computer vision library

%matplotlib inline
```

## Read and show an image


```python
# Read the image
image = mpimg.imread('./pug.jpg')

# Print dimensions
height, width, channels = image.shape
print("Width: {}".format(width))
print("Height: {}".format(height))
print("Depth: {}".format(channels))

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Show grayscale image
plt.imshow(gray_image, cmap="gray")
```

We can also see a sample of the numbers that represent this image:

```python
print(gray_image)
```

We can examine the value of an specific pixel

```python
x = 300
y = 300
print(gray_image[y, x])
```

Given we are dealing with an array of numbers, we can perform math operations, such as finding the min and max:

```python
max_val = np.amax(gray_image)
min_val = np.amin(gray_image)

print("Max: ", max_val)
print("Min: ", min_val)
```

Let's draw a happy face!

```python
tiny_smiley = np.array([[0, 255, 0, 255, 0],
                      [0, 0, 0, 0, 0],
                      [255, 0, 0, 0, 255],
                      [0, 255, 255, 255, 0],
                      [0, 0, 0, 0, 0]])

plt.matshow(tiny_smiley, cmap='gray')
```

## RGB

Up until now we've worked with grayscale images. However, the time has come to work with colors, so we'll explore RGB color space next:

```python
colorful_image = mpimg.imread('./colors.jpeg')

plt.imshow(colorful_image)
```

```python
# Isolate each color channel

red_channel = colorful_image[:, :, 0]
green_channel = colorful_image[:, :, 1]
blue_channel = colorful_image[:, :, 2]

# Now, let's show each channel as a grayscale image. Notice how the color corresponding to each channel appears more intense in the corresponding image
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20, 10))
ax1.set_title("RED channel")
ax1.imshow(red_channel, cmap="gray")

ax2.set_title("GREEN channel")
ax2.imshow(green_channel, cmap="gray")

ax3.set_title("BLUE channel")
ax3.imshow(blue_channel, cmap="gray")
```
